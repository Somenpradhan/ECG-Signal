```python
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np

# Function to plot signal with a label
def plot_signal(time, signal, title, ax):
    ax.plot(time, signal, label=title)
    ax.set_title(title)
    ax.set_xlabel('Time [s]')
    ax.set_ylabel('Amplitude')
    ax.legend()

# Step 1: Read WAV files
file_paths = [
    ('Normal', '/content/drive/MyDrive/signal_analysis/data/N/New_N_001.wav'),
    ('Aortic Stenosis (AS)', '/content/drive/MyDrive/signal_analysis/data/AS/New_AS_001.wav'),
    ('Mitral Stenosis (MS)', '/content/drive/MyDrive/signal_analysis/data/MS/New_MS_003.wav'),
    ('Mitral Valve Prolapse (MVP)', '/content/drive/MyDrive/signal_analysis/data/MVP/New_MVP_005.wav'),
    ('Mitral Regurgitation (MR)', '/content/drive/MyDrive/signal_analysis/data/MR/New_MR_004.wav')
]

# Plot configuration
fig, axes = plt.subplots(len(file_paths), 1, figsize=(10, 12), sharex=True)

# Step 2: Loop through the files, read, and plot
for i, (label, file_path) in enumerate(file_paths):
    signal, sample_rate = sf.read(file_path)
    time = np.linspace(0, len(signal) / sample_rate, num=len(signal))  # Create time axis
    plot_signal(time, signal, f'{label}', axes[i])

# Step 3: Show plot
plt.tight_layout()
plt.show()
```
